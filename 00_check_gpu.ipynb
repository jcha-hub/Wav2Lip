```python
    import subprocess

    try:
        result = subprocess.run(['conda', '--version'], capture_output=True, text=True, check=True)
        print(f"Conda is installed. Version: {result.stdout.strip()}")
    except subprocess.CalledProcessError:
        print("Conda command not found. Conda might not be installed or not in your PATH.")
    except FileNotFoundError:
        print("Conda command not found. Conda might not be installed or not in your PATH.")
```
